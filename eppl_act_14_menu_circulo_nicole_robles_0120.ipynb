Redacción del Problema:Cálculo Geométrico Interactivo: Círculos y Esferas
Descripción:
El objetivo de este programa es desarrollar una herramienta interactiva en Python que permita a los usuarios realizar cálculos geométricos básicos relacionados con el círculo y la esfera. El sistema debe presentar un menú claro que ofrezca cuatro opciones: calcular el área de un círculo, calcular el perímetro (circunferencia) de un círculo, calcular el volumen de una esfera, o salir del programa.
Para lograr esto, se utilizarán estructuras de control de flujo como bucles while para mantener el programa en ejecución hasta que el usuario decida salir, y condicionales if/elif/else para dirigir la elección del usuario a la función correspondiente. Se definirán funciones específicas (areacirculo, perimetrocirculo, volumencirculo) que solicitarán el radio al usuario en cada llamada y realizarán los cálculos utilizando las constantes matemáticas adecuadas. El diseño será amigable, utilizando emojis para mejorar la experiencia visual del usuario.


1. • Área del circulo: A = T. r
2. • Perimetro del circulo: P = 2 • T•r
3. • Volumen de la esfera: V = (4/3) • n• P
* Donde r representa el radio de la figura y r (pi) es la constante matemática aproximadamente igual a 3145965..
* Para implementar estos cálculos de manera práctica, se ha desarrollado un programa en Python estructurado de forma modular.
* El programa utiliza un menú interactivo que se mantiene en ejecución mediante un ciclo whie, permitiendo al usuario realizar múltiples cálculos sin reiniciar la aplicación.
* La navegación se controla con estructuras condicionales if, elif y else, las cuales dirigen el flujo hacia la función corespondiente según la opción elegida.
* Cada operación está encapsulada en su propia función (areacirculo), perimetrocirculo) y volumenesfera(), cumpliendo con el principio de responsabilidad única y facilitando la lectura del código para estudiantes de nivel principiante.
* Al invocar cada función, se solicita directamente el valor del radio, se aplica la fórmula matemática utilizando la librenia estándar math y se muestra el resultado con un formato limpio y decorado con emojis
* Este enfoque no solo refuerza el pensamiento lógico y la sintaxis básica de Python, sino que también demuestra cómo la programación puede servir como herramienta práctica para resolver problemas geométricos de forma rápida y precisa.

```python
import math
#elaborado por Nicole Robles:
# -----------------------------------------------------------------------------
# Definición de Funciones de Cálculo
# -----------------------------------------------------------------------------

def areacirculo():
    """Calcula el área de un círculo solicitando el radio."""
    print("\n--- 🟢 Cálculo del Área del Círculo ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: A = pi * r^2
    area = math.pi * (radio ** 2)

    print(f"✅ El área del círculo con radio {radio} es: {area:.2f}")

def perimetrocirculo():
    """Calcula el perímetro (circunferencia) de un círculo solicitando el radio."""
    print("\n--- 📏 Cálculo del Perímetro del Círculo ---")
    radio = float(input("Por favor, ingresa el radio del círculo: "))

    # Fórmula: P = 2 * pi * r
    perimetro = 2 * math.pi * radio

    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

def volumencirculo():
    """Calcula el volumen de una esfera solicitando el radio."""
    print("\n--- 🔵 Cálculo del Volumen de la Esfera ---")
    radio = float(input("Por favor, ingresa el radio de la esfera: "))

    # Fórmula: V = (4/3) * pi * r^3
    volumen = (4/3) * math.pi * (radio ** 3)

    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

# -----------------------------------------------------------------------------
# Definición de la Función Menú
# -----------------------------------------------------------------------------

def menu():
    """Muestra las opciones disponibles y retorna la elección del usuario."""
    print("\n" + "="*40)
    print("   📐 MENÚ GEOMÉTRICO 📐")
    print("="*40)
    print("1️⃣  Calcular Área del Círculo")
    print("2️⃣  Calcular Perímetro del Círculo")
    print("3️⃣  Calcular Volumen de la Esfera")
    print("4️⃣  Salir del Programa")
    print("="*40)

    opcion = input("👉 Selecciona una opción (1-4): ")
    return opcion

# -----------------------------------------------------------------------------
# Lógica Principal del Programa
# -----------------------------------------------------------------------------

def main():
    """Función principal que controla el flujo del programa."""
    while True:
        # Llamar a la función menú para obtener la elección
        eleccion = menu()

        if eleccion == '1':
            areacirculo()
        elif eleccion == '2':
            perimetrocirculo()
        elif eleccion == '3':
            volumencirculo()
        elif eleccion == '4':
            print("\n👋 ¡Gracias por usar el programa! Hasta luego. 🚀\n")
            break # Sale del bucle while y termina el programa
        else:
            print("\n⚠️  Opción no válida. Por favor, intenta de nuevo.")

# Ejecutar el programa si este archivo es ejecutado directamente
if __name__ == "__main__":
    main()
```
